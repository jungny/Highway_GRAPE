```python
import numpy as np 
import matplotlib.pylab as plt 
import scipy.io

mat_file = scipy.io.loadmat('Data.mat')
```

```python
mat_file['Data'][0][0][30]
```

```python
vel_1 = []
vel_2 = []
vel_3 = []
for i in range(len(mat_file['Data'][0][0])):
    vel_1.append(mat_file['Data'][0][0][i][0][0][3] )
    vel_2.append(mat_file['Data'][0][0][i][1][0][3] )
    vel_3.append(mat_file['Data'][0][0][i][2][0][3] )
```

```python
vel_1
```

```python
dist_list_1 = []
dist_1 = 0

dist_list_2 = []
dist_2 = 0

dist_list_3 = []
dist_3 = 0

for i in range (len(vel_1)):
    dist_1 = vel_1[i] + dist_1
    dist_2 = vel_2[i] + dist_2
    dist_3 = vel_3[i] + dist_3
    dist_list_1.append(dist_1)
    dist_list_2.append(dist_2)
    dist_list_3.append(dist_3)
```

```python
dist_list_1
```

```python
# y_1 = dist_list_1
# y_2 = dist_list_2
# y_3 = dist_list_3

x = np.arange(0, 50.1, 0.1)
```

```python
plt.plot(x,vel_1, label = 'vehicle1')
plt.plot(x,vel_2, label = 'vehicle2')
plt.plot(x,vel_3, label = 'vehicle3')

plt.legend()
plt.xlabel('time(s)')
plt.ylabel('position(m)')
```
